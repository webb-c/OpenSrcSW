```python
# 실습 1 : 서울의 일교차  (210101 ~ 211231)
# 서울에서 일교차가 가장 큰날, 최저기온이 가장 낮은 날, 최고 기온이 가장 높은 날을 출력하라

import csv

# 데이터 읽기
f = open('weather.csv')
data = csv.reader(f)
next(data)

# 일교차, 최저기온, 최고기온 확인 과정
max_temp = -999  # 최고기온
max_date = '' 
min_temp = 100   # 최저기온
min_date = ''
tr_temp = 0     # 일교차
tr_date = ''
for row in data :
    # 누락된 데이터 처리 
    if row[-1] == '' :
        row[-1] = -999
    if row[-2] == '' :
        row[-2] = 100;
        
    # 비교 후 갱신
    row[-1] = float(row[-1])  # 읽어온 데이터는 문자열형태니까 형변환
    row[-2] = float(row[-2])
    tr = row[-1] - row[-2]    # 일교차 : 최고기온 - 최저기온
    if max_temp < row[-1] :   # 저장된 값보다 새로운 데이터가 더 기온이 높으면 갱신
        max_date = row[0]
        max_temp = row[-1]
    if min_temp > row[-2] :
        min_date = row[0]
        min_temp = row[-2]
    if tr_temp < tr :
        tr_date = row[0]
        tr_temp = tr
f.close()

print("서울의 210101 - 211231 관측 데이터 중에서")
print('최고 기온이 가장 높았던 날은', max_date+"입니다.")
print('최저 기온이 가장 낮았던 날은', min_date+"입니다.")
print('일교차가 가장 컸던 날은', tr_date+"입니다.")
```

```python
# 실습 2 : 지하철 이용객
# 노선별로 가장 이용객이 많은 3개 역과 적은 3개 역을 출력하라.

import csv

# 데이터 읽기
f = open('교통카드통계자료.csv')
data = csv.reader(f)
next(data)

# 사용 변수
line = [] #호선 이름 저장
use = [] #이용객 순위 
station = [] #역 순위
people = 0

# 이용객 : 승차 + 하차
for row in data :
    if row[1] not in line:
        # 초기화 전에 역 순위 출력
        if len(line) != 0 :
            # 오름차순 정렬
            for i in range(len(use)):
                for j in range(len(use)-1-i) :
                    if use[j] > use[j+1] :
                        temp = use[j]
                        temp_s = station[j]
                        use[j] = use[j+1]
                        station[j] = station[j+1]
                        use[j+1] = temp
                        station[j+1] = temp_s
            print(line[-1])
            print("이용객이 가장 많은 역 순위")
            for i in range(-1, -4, -1):
                print(station[i], use[i])
            print()
            print("이용객이 가장 적은 역 순위")
            for i in range(3):
                print(station[i], use[i])
            print()
        # 이전에 쓴 배열 초기화
        use.clear()
        station.clear()
        # 새로운 호선 생기면 초기화
        line.append(row[1])
    if row[1] in line:  
        #있는 호선이면 비교해서 갱신
        people = float(row[-1])+float(row[-2])
        use.append(people)
        station.append(row[-3])
        

# 마지막 호선
# 오름차순 정렬
for i in range(len(use)):
    for j in range(len(use)-1-i) :
        if use[j] > use[j+1] :
            temp = use[j]
            temp_s = station[j]
            use[j] = use[j+1]
            station[j] = station[j+1]
            use[j+1] = temp
            station[j+1] = temp_s


print(line[-1])
print("이용객이 가장 많은 역 순위")
for i in range(-1, -4, -1):
    print(station[i], use[i])
print()
print("이용객이 가장 적은 역 순위")
for i in range(3):
    print(station[i], use[i])
```
